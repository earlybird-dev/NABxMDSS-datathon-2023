# Import libraries

```python
import pandas as pd
import numpy as np
```

# Read Data From Excel File

```python
df = pd.read_excel('Data/DATASET.xlsx', 'Database 2022')
df.head()
```

# Data Cleaning

```python
# List all column names
df.columns
```

```python
# Check if column is empty
df[' '].value_counts()
```

```python
# Remove the redundant column
df.drop(columns=[' '], inplace=True)
df.head()
```

```python
# Check column Year unique values
df['Year'].unique()
```

```python
# Remove data of the years 2007-2008, 2011-2012, 2012-2013
df = df[(df['Year'] != '2007-2008')&(df['Year'] != '2011-2012')&(df['Year'] != '2012-2013')]
df['Year'].unique()
```

# Data Transformation

```python
# Row counts of Jurisdiction column
df.Jurisdiction.value_counts()
```

```python
# Convert state/territory short names to the location names recognised by Looker Studio

states = {
          'Tas': 'TAS, Australia', 
          'Vic': 'VIC, Australia', 
          'ACT': 'ACT, Australia', 
          'WA': 'WA, Australia', 
          'SA': 'SA, Australia', 
          'NSW': 'NSW, Australia', 
          'Qld': 'QLD, Australia', 
          'NT': 'NT, Australia', 
          'Australia': 'Australia'
          }

# states = {
#           'Tas': 'Tasmania, Australia', 
#           'Vic': 'Victoria, Australia', 
#           'ACT': 'Australian Capital Territory, Australia', 
#           'WA': 'Western Australia, Australia', 
#           'SA': 'South Australia, Australia', 
#           'NSW': 'New South Wales, Australia', 
#           'Qld': 'Queensland, Australia', 
#           'NT': 'Northern Territory, Australia', 
#           'Australia': 'Australia'
#           }

df['State'] = df['Jurisdiction'].apply(lambda x: states[x])
df.head()
```

## Extract Data By Classification (i.e.'Total', 'Category', 'Type')

```python
# Filter out the redundant rows with Jurisdiction = 'Australia' & 'Stream' = 'Total'
df_state = df[(df['Jurisdiction'] != 'Australia')&(df['Stream'] != 'Total')]
df_state.head()
```

```python
# Extract rows with 'Total' Classification
df_classification_total = df_state[df_state['Classification'] == 'Total']
df_classification_total.to_csv('Data/df_classification_total.csv', index=False)
```

```python
# Extract rows with 'Category' Classification
df_classification_category = df_state[df_state['Classification'] == 'Category']
df_classification_category.to_csv('Data/df_classification_category.csv', index=False)
```

```python
# Extract rows with 'Type' Classification
df_classification_type = df_state[df_state['Classification'] == 'Type']
df_classification_type.to_csv('Data/df_classification_type.csv', index=False)
```

# Extract Recovery Rate Data

## Recovery Rate By Year

```python
# Filter out recovery and recycling data rows
df_recovery = df_classification_type[(df_classification_type['Fate'] == 'Recycling') | (df_classification_type['Fate'] == 'Waste reuse') | (df_classification_type['Fate'] == 'Energy recovery')]
df_recycling = df_classification_type[(df_classification_type['Fate'] == 'Recycling')]
```

```python
# Total waste by year
df_waste_recovered_by_year = df_classification_type.groupby('Year').sum('Tonne')
df_waste_recovered_by_year.reset_index(inplace=True)
df_waste_recovered_by_year
```

```python
# Waste recovered by year
df_waste_recovered_groupby = df_recovery.groupby('Year').sum('Tonne')
df_waste_recovered_groupby.reset_index(inplace=True)
df_waste_recovered_groupby
```

```python
# Waste recycled by year
df_waste_recycled_groupby = df_recycling.groupby('Year').sum('Tonne')
df_waste_recycled_groupby.reset_index(inplace=True)
df_waste_recycled_groupby
```

```python
# Join data with this method as tables have the same size and index
df_waste_recovered_by_year['Recovery'] = df_waste_recovered_groupby['Tonnes']
df_waste_recovered_by_year['Recycling'] = df_waste_recycled_groupby['Tonnes']
df_waste_recovered_by_year['Not Recovery'] = df_waste_recovered_by_year['Tonnes'] - df_waste_recovered_by_year['Recovery']
df_waste_recovered_by_year['Recovery Rate'] = df_waste_recovered_by_year['Recovery'] / df_waste_recovered_by_year['Tonnes']
df_waste_recovered_by_year['Recycling Rate'] = df_waste_recovered_by_year['Recycling'] / df_waste_recovered_by_year['Tonnes']
df_waste_recovered_by_year
```

```python
# Export data
df_waste_recovered_by_year.to_csv('Data/df_waste_recovered_by_year.csv', index=False)
```

## Recovery Rate By Year and Jurisdiction

```python
# Total waste by year and state
df_waste_recovered_by_year_state = df_classification_type.groupby(['Year','Jurisdiction']).sum('Tonne')
df_waste_recovered_by_year_state.reset_index(inplace=True)
df_waste_recovered_by_year_state
```

```python
# Waste recovered
df_waste_recovered = df_recovery.groupby(['Year','Jurisdiction']).sum('Tonne')
df_waste_recovered.reset_index(inplace=True)
df_waste_recovered
```

```python
# Waste recycled
df_waste_recycled = df_recycling.groupby(['Year','Jurisdiction']).sum('Tonne')
df_waste_recycled.reset_index(inplace=True)
df_waste_recycled
```

```python
# Join data with this method as tables have the same size and index
df_waste_recovered_by_year_state['Recovery'] = df_waste_recovered['Tonnes']
df_waste_recovered_by_year_state['Recycling'] = df_waste_recycled['Tonnes']
df_waste_recovered_by_year_state['Not Recovery'] = df_waste_recovered_by_year_state['Tonnes'] - df_waste_recovered_by_year_state['Recovery']
df_waste_recovered_by_year_state['Recovery Rate'] = df_waste_recovered_by_year_state['Recovery'] / df_waste_recovered_by_year_state['Tonnes']
df_waste_recovered_by_year_state['Recycling Rate'] = df_waste_recovered_by_year_state['Recycling'] / df_waste_recovered_by_year_state['Tonnes']
df_waste_recovered_by_year_state
```

```python
# Export data
df_waste_recovered_by_year_state.to_csv('Data/df_waste_recovered_by_year_state.csv', index=False)
```
